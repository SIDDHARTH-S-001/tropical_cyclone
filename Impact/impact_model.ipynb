```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential, load_model
from keras.layers import Dense
import numpy as np
```

```python
# read the CSV file
df = pd.read_csv('impact_data.csv')

# extract input and output columns
input_cols = df.iloc[:, :4] # extract the first four columns
output_cols = df.iloc[:, -1] # extract the last column
```

```python
output_cols
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_cols, output_cols, test_size=0.25)
```

```python
impact_model = Sequential()
impact_model.add(Dense(4, input_dim=4, activation='relu'))
impact_model.add(Dense(15, activation='relu'))
impact_model.add(Dense(1, activation='linear'))
```

```python
impact_model.compile(
   loss = 'mae', optimizer = 'adam', metrics = ['mean_squared_error']
)
```

```python
impact_model.fit(X_train, y_train, epochs=1500, batch_size=4, verbose=1)
```

```python
impact_model.summary()
```

```python
impact_model.save("impact_1.h5")
```

```python
model = load_model('impact_1.h5')
```
